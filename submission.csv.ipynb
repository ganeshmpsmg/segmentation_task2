```python
import os
import pandas as pd
# Search for Mall_Customers.csv on your system
search_drives = ['C:\\', 'D:\\']
found_paths = []

for drive in search_drives:
    for root, dirs, files in os.walk(drive):
        if 'Mall_Customers.csv' in files:
            found_paths.append(os.path.join(root, 'Mall_Customers.csv'))
            break  # stop after first find per drive

if found_paths:
    print('✅ Found at:', found_paths[0])
    df = pd.read_csv(found_paths[0])
    print('Shape:', df.shape)
else:
    print('❌ Not found — please download from Kaggle first')
    print('🔗 https://www.kaggle.com/datasets/vjchoudhary7/customer-segmentation-tutorial-in-python')
```
